```python
writePath = r'D:\Final Year Project\Data\SiTa\Extracted'
for file in files:
  fileName, extension = splitext(file)
  raw_data = pd.read_excel(join(path, file), header=None)
  try:
    data = raw_data[0]
    newData = []
    for i in range(len(data)):
      try:
        x = data[i].strip().split()
      except:
        continue
      newData += x
    # newData = sorted(list(set(newData)))
    with open(join(writePath, f'{fileName}.txt'), 'w' , encoding='utf-8') as f:
      f.write('\n'.join(newData))
  except Exception as e:
    print('Error occured in ' + fileName)
    print(e)
```

```python
path = r'D:\Final Year Project\Data\SiTa\Extracted'
files = [f for f in listdir(path) if isfile(join(path, f))]
```

```python
names = []
for file in files:
  fileName, extension = splitext(file)
  raw_data =  open(join(path, file), encoding='utf-8').read()
  names += raw_data.split('\n')
```

```python
from unicodedata import normalize
names = [normalize('NFD', name.strip()) for name in names]
uniqueNames = sorted(list(set(names)))
```

```python
len(uniqueNames)
```

```python
writePath = r'D:\Final Year Project\Data\SiTa\Extracted\Combined'
with open(join(writePath, 'combinedUniqueNames.txt'), 'w', encoding='utf-8') as f:
  f.write('\n'.join(uniqueNames))

with open(join(writePath, 'combinedAllNames.txt'), 'w', encoding='utf-8') as f:
  f.write('\n'.join(names))
```

```python
filePath = r'D:\Final Year Project\Data\SiTa\Extracted\Combined\combinedAllNames.txt'
names = open(filePath, 'r', encoding='utf-8').readlines()
cleanedNames = []
for line in names:
    line = line.replace('fප', 'ෆ')
    line = line.replace('Fප', 'ෆ')
    for name in line.split('.'):
        cleanedName = ''
        for character in name:
            cleanedName += character if( 3456 < ord(character) < 3583 or character == '\u200d' ) else ''
        if(len(cleanedName) > 2):
            cleanedNames.append(cleanedName)
```

```python
filePath = r'D:\Final Year Project\Data\SiTa\Extracted\Combined\combinedAllNames- Cleaned.txt'
open(filePath, 'w', encoding='utf-8').write('\n'.join(cleanedNames))

filePath = r'D:\Final Year Project\Data\SiTa\Extracted\Combined\combinedUniqueNames- Cleaned.txt'
open(filePath, 'w', encoding='utf-8').write('\n'.join(list(set(cleanedNames))))
```
